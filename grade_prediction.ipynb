```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn import linear_model
pd.__version__






```

```python
loan_2015 = pd.read_csv('data/2015.csv')

y = loan_2015[~loan_2015.grade.isnull()].grade
feature_list = ['loan_amnt', 'term', 
       'emp_length', 'home_ownership', 'annual_inc',
       'pymnt_plan','purpose', 'dti', 'delinq_2yrs', 'inq_last_6mths',
        'open_acc','pub_rec', 'revol_bal', 'revol_util', 'total_acc',
       'acc_now_delinq','open_acc_6m',
       'open_il_6m', 'open_il_12m', 'open_il_24m','total_bal_il', 'il_util', 'open_rv_12m', 'open_rv_24m',
       'max_bal_bc', 'all_util', 'total_rev_hi_lim', 
       'total_cu_tl', 'inq_last_12m']
X = pd.get_dummies(loan_2015.ix[~loan_2015.grade.isnull(), feature_list])



```

```python
X.fillna(0, inplace=True)

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

```python
logreg = linear_model.LogisticRegression(C=1e5)

logreg.fit(X_train, y_train)


logreg.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

pred = logreg.predict(X_test)

confusion_matrix(y_test, pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
 
```

```python
rf.score(X_test, y_test)
```

```python
pred = rf.predict(X_test)
confusion_matrix(y_test, pred)
```
